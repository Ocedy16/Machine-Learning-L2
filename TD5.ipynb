```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
X = data.data
y = data.target
print(X.shape)

#Il y a 569 exemples et 30 variables descriptives
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y)

y_train.size
y_test.size
```

```python
from sklearn.linear_model import Perceptron
perceptron = Perceptron()
perceptron.fit(X_train,y_train)
perceptron.score(X_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import validation_curve
k_range=range(1,21)
#knn=KNeighborsClassifier(n_neighbors=k_range)
#scores=cross_val_score(knn,X_train,y_train,cv=7)
train_scores, valid_scores = validation_curve(KNeighborsClassifier(),
                                              X_train,
                                              y_train,
                                              param_name="n_neighbors",
                                              param_range=k_range,
                                              cv=7)

train_mean_score = np.mean(train_scores,axis=1)
valid_mean_score = np.mean(valid_scores,axis=1)

plt.plot(k_range, train_mean_score, label='Score d\'apprentissage')
plt.plot(k_range, valid_mean_score, label='Score de validation')
plt.xlabel("Hyperparam√®tre k")
plt.ylabel("Score 7-CV")
plt.legend()
plt.show()
```

```python
k_best = k_range[np.argmax(valid_mean_score)]
#k_best=14
knn=KNeighborsClassifier(n_neighbors=k_best)
knn.fit(X_train,y_train)
knn.score(X_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression(C=140,penalty="l1",solver="liblinear")
logreg.fit(X_train,y_train)
print(logreg.score(X_test,y_test))
```

```python
logreg=LogisticRegression(C=array,penalty="l1",solver="liblinear")
train_scores, valid_scores = validation_curve(logreg,
                                              X_train,
                                              y_train,
                                              param_name="C",
                                              param_range=array,
                                              cv=7)

train_mean_score = np.mean(train_scores,axis=1)
valid_mean_score = np.mean(valid_scores,axis=1)

plt.plot(np.log10(array), train_mean_score, label='Score d\'apprentissage')
plt.plot(np.log10(array), valid_mean_score, label='Score de validation')
plt.xlabel("Hyperparam√®tre k")
plt.ylabel("Score 7-CV")
plt.legend()
plt.show()
```

```python
C_best = array[np.argmax(valid_mean_score)]

logreg = LogisticRegression(penalty='l1',solver='liblinear',C=C_best).fit(X_train,y_train)
print(logreg.score(X_test,y_test))
```
