```python
import numpy as np
from sklearn.datasets import fetch_california_housing
data = fetch_california_housing()
X = data.data
y = data.target
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=17)
data.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(random_state=0).fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
def tree_summary (dt):
  train_score = dt.score(X_train,y_train)
  y_predict = dt.predict(X_test)
  test_score = r2_score(y_predict, y_test)
  profondeur = dt.tree_.max_depth
  nb_noeuds = dt.tree_.node_count
  return train_score, test_score, profondeur, nb_noeuds
tree_summary(dt)
```

```python
dt2 = DecisionTreeRegressor (random_state = 0, max_depth = 3).fit(X_train, y_train)
print(tree_summary (dt2))
from sklearn.tree import plot_tree
plot_tree(dt2,fontsize=8, label=None, impurity=False)
```
