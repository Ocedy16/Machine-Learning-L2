```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('https://joon-kwon.github.io/cpes-ml/wine_dataset.csv')
data.head()

X = data[['chlorides','total_sulfur_dioxide']]
y = data['style']

from sklearn.preprocessing import MinMaxScaler
X = MinMaxScaler(feature_range=(-1,1)).fit(X).transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
from sklearn.linear_model import LogisticRegressionCV
C_range=np.logspace(-6,6,20)
logreg=LogisticRegressionCV(Cs=C_range,cv=5)
logreg.fit(X_train,y_train)
logreg.score(X_test,y_test)
```

```python
def plot_decision_boundary(predictor):
  plt.scatter(X[:,0],X[:,1],c=y,edgecolors='k')
  figure = plt.gca()
  xlim = figure.get_xlim()
  ylim = figure.get_ylim()
  xx = np.linspace(xlim[0], xlim[1], 100)
  yy = np.linspace(ylim[0], ylim[1], 100)
  YY, XX = np.meshgrid(yy, xx)
  xy = np.vstack([XX.ravel(), YY.ravel()]).T
  P = predictor.decision_function(xy).reshape(XX.shape)
  plt.contour(XX, YY, P, levels=[0], alpha=1)
  plt.show()
plot_decision_boundary(logreg)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import LinearSVC
grid = {'C':np.logspace(-6,6,20)}
svc =GridSearchCV(LinearSVC(),cv=5,param_grid=grid,verbose=3)
print('Score:',svc.fit(X_train,y_train).score(X_test,y_test))
plot_decision_boundary(svc)
print(svc.best_params_)
```

```python
from sklearn.svm import SVC
liste = []
d_range = range(2,9)
for deg in range (2,9):
  svcpoly = SVC(kernel='poly',C=1,coef0=1,degree=deg,gamma='auto')
  svcpoly.fit(X_train,y_train)
  plot_decision_boundary(svcpoly)
```

```python
from sklearn.svm import LinearSVC
param_grid = {'degree': range(2, 9)}
svc = GridSearchCV(SVC(kernel='poly', C=1, coef0=1, gamma='auto'), cv=5, param_grid=param_grid, verbose=3)
print('Score', svc.fit(X_train, y_train).score(X_test, y_test))
print(svc.best_params_)


svcpoly_best = SVC(kernel='poly', C=1, coef0=1, degree=svc.best_params_['degree'], gamma='auto')
svcpoly_best.fit(X_train, y_train)
print('Score', svcpoly_best.fit(X_train, y_train).score(X_test, y_test))
plot_decision_boundary(svcpoly_best)
```
